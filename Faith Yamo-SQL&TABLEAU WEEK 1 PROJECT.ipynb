Write a SQL script that creates the following summary tables
a) District_summary.

 i) District Name

```python
%%sql
SELECT jurisdiction, jurisdiction_type
FROM jurisdiction_hierarchy
WHERE jurisdiction_type = 'District'
```

ii) Region Name

```python
%%sql
SELECT jurisdiction, jurisdiction_type
FROM jurisdiction_hierarchy
WHERE jurisdiction_type = 'Region'
```

iii) No. of Individual Beneficiaries (1HH = 6 Individuals)

```python
%%sql
SELECT SUM(beneficiaries)*6 AS beneficiaries, beneficiary_type
FROM beneficiary_partner_data
WHERE beneficiary_type = 'Households'
```

```python
%%sql
SELECT SUM(beneficiaries) AS beneficiaries, beneficiary_type
FROM beneficiary_partner_data
WHERE beneficiary_type = 'Individuals'
```

```python
%%sql
SELECT SUM(
  CASE
   WHEN beneficiary_type = 'Individuals' THEN beneficiaries
   WHEN beneficiary_type = 'Households' THEN beneficiaries * 6
   ELSE 0
   END ) AS individualbeneficiaries
FROM beneficiary_partner_data;
```

iv) No. of Individual Beneficiaries / Total District Population

```python
%%sql
WITH DistrictBeneficiaries AS (
SELECT JH.parent_jurisdiction AS district,
SUM(
  CASE
   WHEN beneficiary_type = 'Individuals' THEN beneficiaries
   WHEN beneficiary_type = 'Households' THEN beneficiaries * 6
   ELSE 0
   END ) AS individualbeneficiaries
FROM beneficiary_partner_data BP
JOIN jurisdiction_hierarchy JH
ON BP.village = JH.jurisdiction
GROUP BY JH.parent_jurisdiction
),
DistrictPopulation AS (
  SELECT
  JH.parent_jurisdiction AS district, SUM(VL.total_population) AS totalpopulation
  FROM village_locations VL
  JOIN jurisdiction_hierarchy JH
  ON VL.village = JH.jurisdiction
  GROUP BY JH.parent_jurisdiction
)
SELECT DB.district, DB.individualbeneficiaries, (DB.individualbeneficiaries) / DP.totalpopulation AS ratio
FROM DistrictBeneficiaries DB
JOIN DistrictPopulation DP
ON DB.district = DP.district
```

 Partner_summary

i) Partner Name

```python
%%sql
SELECT DISTINCT(partner) AS partnername
FROM beneficiary_partner_data
```

ii) No. of Villages reached by partner

```python
%%sql
SELECT partner, COUNT(DISTINCT village) AS villages_reached
FROM beneficiary_partner_data BP
GROUP BY partner
```

iii) No. of Districts reached by partner

```python
%%sql
SELECT BP.partner, COUNT( DISTINCT JH.parent_jurisdiction) AS Districtsreached
FROM beneficiary_partner_data BP
JOIN jurisdiction_hierarchy JH
ON BP.village = JH.jurisdiction
GROUP BY BP.partner
```
