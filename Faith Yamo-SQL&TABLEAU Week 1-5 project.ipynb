WEEK 4



Which customer has made the most orders?


```python
%%sql
SELECT O.customer_id, C.first_name, COUNT(O.order_id) AS ordersmade
FROM orders O
join cutomers C
on O.customer_id = C.customer_id
GROUP BY C.first_name
ORDER BY ordersmade DESC
```

What is the total revenue per product


```python
%%sql
SELECT P.product_name,
 P.price * Sum(OI.quantity) AS totalrevenue
FROM products P
join order_items OI
on P.product_id = OI.product_id
GROUP BY P.product_name, P.price
```

WEEK 5


Find the first order(by date) for each customer

```python
%%sql
SELECT C.customer_id,C.first_name, O.order_id, O.order_date
FROM cutomers C
join orders O
on C.customer_id = O.customer_id
```

What's the total revenue per product

```python
%%sql
SELECT P.product_name,
SUM(P.price * OI.quantity) AS totalrevenue
FROM products P
join order_items OI
on P.product_id = OI.product_id
GROUP BY P.product_name, P.price
```

Find the day with the highest revnue

```python
%%sql
SELECT O.order_date,
SUM(P.price * OI.quantity) AS totalrevenue
FROM products P
join order_items OI
on P.product_id = OI.product_id
join orders O
on O.order_id = OI.order_id
GROUP BY O.order_date
ORDER BY totalrevenue DESC
LIMIT 1;
```
