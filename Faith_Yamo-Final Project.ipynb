```python
%%sql DROP TABLE IF EXISTS Regions;
```

```python
%%sql CREATE TABLE Regions
(Region_ID INTEGER PRIMARY KEY,
 Region_Name Varchar (20),
 Town VARCHAR (20));
```

```python
%%sql DROP TABLE IF EXISTS Prices;
```

```python
%%sql CREATE TABLE Prices
(Date TEXT,
 Petrol_Price decimal,
 Diesel_Price decimal,
 Kerosene_Price decimal,
 Town varchar,
 FOREIGN KEY (Town) REFERENCES Regions(Town));
```
